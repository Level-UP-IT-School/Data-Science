```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 * np.sin(x)
```

```python
def f_dash(x):
    return 2*x*np.sin(x)+x**2*np.cos(x)
```

```python
def draw_plot(f, x_begin, x_end):
    x_array = np.linspace(x_begin, x_end, 1000)
    plt.plot(x_array, f(x_array))
```

```python
draw_plot(f, 1, 8)
```

```python
def gradient_descent_step(f_dash, x, alpha = 0.001):
    f_dash_x=f_dash(x)
    delta_x=alpha*f_dash_x
    x_new= x- delta_x
    return x_new
```

```python
gradient_descent_step(f_dash, 4)
```

```python
def gradient_descent(f, f_dash, x, alpha=0.001, epsilon = 0.01):
    
    y=f(x)
    xs = [x]
    ys = [y]
    
    while True:
        x = gradient_descent_step(f_dash, x, alpha)
        y = f(x)
        previous_y = ys[-1]
        delta_y = np.abs(y-previous_y)
        xs.append(x)
        ys.append(y)
        print('Текущий x: {}, текущий y: {}'.format(x,y))
        print("Рфзница двух последних y: {}".format(delta_y))
        
        if delta_y < epsilon:
            break
    
    return xs, ys
    
```

```python
def draw_plot(f, x_begin, x_end, xs, ys):
    x_array = np.linspace(x_begin, x_end, 1000)
    plt.plot(x_array, f(x_array))
    
    plt.scatter(xs, ys, c='r')
    
    plt.annotate(' x = {}, y = {}'.format(round(xs[-1], 1), round(ys[-1], 1)), (xs[-1], ys[-1]))
    
    plt.title("f(x)")
    plt.show()
```

```python
alpha = 0.0005
epsilon = 0.0001
```

```python
x = 2.5
```

```python
xs, ys = gradient_descent( f, f_dash, x, alpha, epsilon)
```

```python
draw_plot(f, x_begin=1, x_end=8, xs=xs, ys=ys)
```
